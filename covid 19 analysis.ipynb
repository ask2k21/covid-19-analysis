```python
# import of libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import of datasets

data = pd.read_csv("covid19_Confirmed_dataset.csv")

data.head()
```

```python
data.shape
```

```python
# deleting unnecessary columns
data.drop(["Lat","Long"],axis =1,inplace = True)
```

```python
data.head()
```

```python
data.shape
```

```python
corona_data_aggregated = data.groupby("Country/Region").sum()
```

```python
corona_data_aggregated.head()
```

```python
corona_data_aggregated.shape
```

```python
# visualize data for countries

corona_data_aggregated.loc["China"].plot()
corona_data_aggregated.loc["India"].plot()
corona_data_aggregated.loc["Pakistan"].plot()

plt.legend()
```

```python
corona_data_aggregated.loc["China"].diff().plot()
```

```python
# maximum infection rate
corona_data_aggregated.loc["China"].diff().max()
```

```python
countries = corona_data_aggregated.index

max_infection_rate = []

for i in countries:
    max_infection_rate.append(corona_data_aggregated.loc[i].diff().max())
corona_data_aggregated["Max_infection_rate"] = max_infection_rate
```

```python
corona_data_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_data_aggregated["Max_infection_rate"])
```

```python
corona_data
```

```python
# importing worldhappiness report

happiness_report = pd.read_csv("worldwide_happiness_report.csv")

happiness_report.head()
```

```python
# drop useless columns
list = ["Overall rank","Score","Generosity","Perceptions of corruption"]

happiness_report.drop(list,axis =1,inplace = True)
```

```python
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

```python
data1 = corona_data.join(happiness_report, how = "inner")

data1
```
